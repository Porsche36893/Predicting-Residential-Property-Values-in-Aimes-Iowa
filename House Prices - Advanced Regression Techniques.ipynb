### Finding Missing Values
- Print percentage of missing value in each columns

```python
# train_data.isnull() returns pandas DataFrame with same shape as train_data but have boolean True/False in each cell where True = NaN (missing value)
# train_data.isnull().sum() returns pandas Series where each element represents total number of observations with missing value in that column
# train_data.shape[0] returns number of observation in train_data
missing_col = train_data.columns[train_data.isnull().sum()/train_data.shape[0]*100>0] # returns list of columns with missing value
missing_col = pd.DataFrame({'Columns':missing_col,
              'Percentage':train_data[missing_col].isnull().sum()/train_data.shape[0]*100})
missing_col.reset_index(drop=True,inplace=True)
missing_col.sort_values(by='Percentage',ascending=False,inplace=True)
missing_col
```

#### Let's see the missing value correlation with target feature
- Observing how missing value correlated with target feature
- Using bar chart which 1 represent the missing value and 0 for non
missing vlaue

```python
for col in missing_col.Columns:
    data = train_data.copy() # copy DataFrame to another variable (unable to do data = train_data because then data will refer to train_data memory)
    data[col] = np.where(data[col].isnull() == True, 1, 0) # replace value of column col by 1 if there is missing value else 0
    data.groupby(col)['SalePrice'].mean().plot.bar() # group data by value 0, 1 in column col and compute mean in each group and plot barplot
    plt.title(col) # specify title of plot to be col
    plt.show()
```

###### Observtaion :
* Alley, MasVnrType, MasVnrArea, Fence, MiscFeature has slighlty high SalePrice  for the missing value
* BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2, FireplaceQu, GarageType, GarageYrBlt, GarageFinish, GarageQual, GarageCond, PoolQC has less SalePrice for missing value

# **Hyperparameter Tuning using BayesSearchCV**

The reason why we try to use Bayesian Optimization to find the optimized hyperparameter is due to its nature for finding hyperparameter point that maximizes the reward (negative of loss).

The algorithm runs as followed: <br>
1. Sample a hyperparameter point that one wants to tune
2. Perform cross-validation on model with sampled hyperparameter and get the loss value of selected metric (MAE) and compare it with the best hyperparameter point stored
3. Use hyperparameters as features and loss value as target to estimate the loss value statistics (mean and variance) over other possible hyperparameters' value. (This is where Bayesian/Gaussian Process is used; Surrogate Model)
4. Choose the next hyperparameter point such that it (will likely) maximizes the negative loss value of model with chosen hyperparameter point. (This is where Optimization is used; Acquisiton Function; choose point with high mean = exploitation, choose point with high variance = exploration)
5. From sampled point in 4, repeat 2-4 until maximum iteration is hit.

```python
# Importing important libraries
from skopt import BayesSearchCV # skopt is an equivalent name for scikit-optimize
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
```

### Tuning Hyperparameters of XGBRegressor
Hyperparameter definition :     
1. model__n_estimators:

- Controls the number of boosting rounds, essentially the number of trees in the ensemble.
- More trees generally lead to better performance but can also increase training time and risk overfitting.
2. model__learning_rate:

- Determines the step size at each iteration.
- Smaller values often require more iterations but can lead to more stable and robust models.
- The 'log-uniform' distribution suggests a logarithmic scale for this parameter.
3. model__max_depth:

- Sets the maximum depth of each tree.
- Deeper trees can capture complex patterns but are more prone to overfitting.
4. model__subsample:

- Specifies the fraction of training data used for each tree.
- Lower values can reduce overfitting but might also reduce model performance.
5. model__colsample_bytree:

- Determines the fraction of features used for each tree.
- Similar to subsample, it can help prevent overfitting.
6. model__gamma:

- Sets the minimum loss reduction required to make a split.
- Higher values can lead to simpler models.
7. model__reg_alpha (L1 regularization):

- Adds an L1 penalty to the model's weights.
- Can help prevent overfitting and feature selection.
8. model__reg_lambda (L2 regularization):

- Adds an L2 penalty to the model's weights.
- Can help prevent overfitting and improve model generalization.

```python
# Define the XGBRegressor model and pipeline
model = XGBRegressor()
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])

# Define the parameter search space for XGBRegressor
param_space = {
    'model__n_estimators': (50, 500),                    # Number of boosting rounds
    'model__learning_rate': (0.01, 0.3, 'log-uniform'),  # Learning rate
    'model__max_depth': (3, 10),                         # Maximum tree depth
    'model__subsample': (0.5, 1.0),                      # Subsample ratio of the training instance
    'model__colsample_bytree': (0.5, 1.0),               # Subsample ratio of columns
    'model__gamma': (0, 5),                              # Minimum loss reduction to make a split
    'model__reg_alpha': (1e-5, 1e1, 'log-uniform'),      # L1 regularization term on weights
    'model__reg_lambda': (1e-5, 1e1, 'log-uniform')      # L2 regularization term on weights
}

# Initialize BayesSearchCV with the pipeline, search space, and scoring metric
bayes_cv_tuner = BayesSearchCV(
    estimator=pipeline,
    search_spaces=param_space,
    n_iter=30,  # Number of parameter settings sampled
    scoring='neg_mean_absolute_error',
    cv=10,
    n_jobs=-1,  # Use all available cores
    verbose=0,
    random_state=0
)

# Run Bayesian Optimization
bayes_cv_tuner.fit(X_train, y_train)

# Get the best score and parameters
best_score = -bayes_cv_tuner.best_score_
best_params = bayes_cv_tuner.best_params_

print("Best Mean Absolute Error:", best_score)
print("Best Parameters:", best_params)
```

### Tuning Hyperparameters of RandomForestRegressor
Hyperparamter definition :     
1. model__n_estimators:

- Controls the number of trees in the forest.
- More trees generally lead to better performance but can also increase training time.
2. model__max_depth:

- Sets the maximum depth of each tree.
- Deeper trees can capture complex patterns but are more prone to overfitting.
- None means the trees will grow until all leaves are pure.
3. model__min_samples_split:

- Determines the minimum number of samples required to split an internal node.
- Larger values can lead to simpler trees.
4. model__min_samples_leaf:

- Specifies the minimum number of samples required to be at a leaf node.
- Larger values can lead to more conservative models.
5. model__max_features:

- Controls the number of features to consider when looking for the best split.
'sqrt' and 'log2' are common choices,
6. model__bootstrap:

- Determines whether bootstrap samples are used when building trees.
- Bootstrapping can help reduce variance.

```python
# Define the RandomForestRegressor model and pipeline
model = RandomForestRegressor()
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])

# Define the parameter search space for RandomForestRegressor
param_space = {
    'model__n_estimators': (50, 500),                    # Number of trees in the forest
    'model__max_depth': (None, 5, 10, 15, 20),           # Maximum depth of the trees (None means nodes are expanded until all leaves are pure)
    'model__min_samples_split': (2, 10),                  # Minimum number of samples required to split an internal node
    'model__min_samples_leaf': (1, 10),                   # Minimum number of samples required to be at a leaf node
    'model__max_features': ['sqrt', 'log2'],     # Number of features to consider when looking for the best split
    'model__bootstrap': [True, False],                    # Whether bootstrap samples are used when building trees
}

# Initialize BayesSearchCV with the pipeline, search space, and scoring metric
bayes_cv_tuner = BayesSearchCV(
    estimator=pipeline,
    search_spaces=param_space,
    n_iter=30,  # Number of parameter settings sampled
    scoring='neg_mean_absolute_error',
    cv=10,
    n_jobs=-1,  # Use all available cores
    verbose=0,
    random_state=0
)

# Run Bayesian Optimization
bayes_cv_tuner.fit(X_train, y_train)

# Get the best score and parameters
best_score = -bayes_cv_tuner.best_score_
best_params = bayes_cv_tuner.best_params_

print("Best Mean Absolute Error:", best_score)
print("Best Parameters:", best_params)
```

### Tuning Hyperparameters of Ridge Regression
Hyperparamter definition :
1. model__alpha:

- Controls the strength of the L2 regularization term.
- Higher values of alpha lead to stronger regularization, which can help prevent overfitting, especially in high-dimensional datasets.
- The 'log-uniform' distribution suggests a logarithmic scale for this parameter.
2. model__solver:
- Specifies the algorithm used to solve the optimization problem.

3. model__fit_intercept:
- Determines whether to calculate the intercept term in the model.

4. model__max_iter:

- Sets the maximum number of iterations for solvers that support iterative optimization.
- A higher value can improve accuracy but may increase computation time.

```python
# Define the Ridge Regression model and pipeline
model = Ridge()
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])

# Define the parameter search space for Ridge regression
param_space = {
    'model__alpha': (1e-5, 1e2, 'log-uniform'),           # Regularization strength
    'model__solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sag', 'saga'],  # Algorithm to use for optimization
    'model__fit_intercept': [True, False],                # Whether to fit the intercept or not
    'model__max_iter': (100, 1000),                       # Maximum number of iterations for solvers that support it
}

# Initialize BayesSearchCV with the pipeline, search space, and scoring metric
bayes_cv_tuner = BayesSearchCV(
    estimator=pipeline,
    search_spaces=param_space,
    n_iter=30,  # Number of parameter settings sampled
    scoring='neg_mean_absolute_error',
    cv=10,
    n_jobs=-1,  # Use all available cores
    verbose=0,
    random_state=0
)

# Run Bayesian Optimization
bayes_cv_tuner.fit(X_train, y_train)

# Get the best score and parameters
best_score = -bayes_cv_tuner.best_score_
best_params = bayes_cv_tuner.best_params_

print("Best Mean Absolute Error:", best_score)
print("Best Parameters:", best_params)
```
